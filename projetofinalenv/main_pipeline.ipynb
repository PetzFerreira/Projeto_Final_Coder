- Importação de bibliotecas

```python
import pandas as pd
import requests
import sqlite3 as sql
from plyer import notification
import datetime
```

- Função de disparo de alerta

```python
def alert(nivel, base, response):
    alerta = 'Atenção!'
    message = 'Código API retornado: '
    match nivel:
        case 1:
            nivel = 'Baixo'
        case 2:
            nivel = 'Moderado'
        case 3:
            nivel = 'Alto'
    try:
        notification.notify(title= f'{alerta.upper()}', message = f'Falha na base {base}, Nível {nivel} - {message} {response}', timeout=15)  # type: ignore

    except Exception as e:
        print(f'Erro {e}')
```

- Função para confirmar código API

```python
'''Essa função retorna o código api 200 ("OK", ou "Requisição bem sucedida") ou
retorna notificação apresentando o código adquirido.'''

def Confirmar_resposta_api(url):
    response = requests.get(url)
    
    if response.status_code == 200:    
        return response
   
    else:
        alert(1, url, response.status_code)
        print(f'Falha na obtenção dos dados solicitados. código do erro: {response.status_code}')
               
```

-Função que cria dataframe completo a partir do status_code:

```python
''' Essa função '''

def criar_dataframe(resp):
   resp_data = resp.json()
   resp_data_df = pd.DataFrame(resp_data)
   return resp_data_df
```

- Função para verificação (Head())

```python
''' Essa função '''

def ler_head_url(resp):
   resp_data = resp.json()
   resp_data_df = pd.DataFrame(resp_data)
   return resp_data_df.head()
```

- Função para verificação (Tail())

```python
def ler_tail_url(resp):
    resp_data = resp.json()
    resp_data_df = pd.DataFrame(resp_data)
    return resp_data_df.tail()
```

- Função que retorna Dataframe (Head) com cidades e estados de acordo com o DDD desejado:

```python
def obter_dados_do_ddd_head(ddd):
    url_ddd = f'https://brasilapi.com.br/api/ddd/v1/{ddd}'
    response = requests.get(url_ddd)
    
    if response.status_code == 200:
        response = response.json()
        df_ddd = pd.DataFrame(response)
        df_ddd['ddd'] = ddd
        return df_ddd.head()
    else:
        alert(1, url_ddd, response.status_code)
        print(f'Falha na obtenção dos dados solicitados. código do erro: {response.status_code}')
```

- Função que retorna Dataframe (Tail) com cidades e estados de acordo com o DDD desejado:

```python
def obter_dados_do_ddd_tail(ddd):
    url_ddd = f'https://brasilapi.com.br/api/ddd/v1/{ddd}'
    response = requests.get(url_ddd)
    
    if response.status_code == 200:
        response = response.json()
        df_ddd = pd.DataFrame(response)
        df_ddd['ddd'] = ddd
        return df_ddd.tail()
    else:
        alert(1, url_ddd, response.status_code)
        print(f'Falha na obtenção dos dados solicitados. código do erro: {response.status_code}')
```

- Função que retorna Dataframe completo com cidades e estados de acordo com o DDD desejado:

```python
def dataframe_completa_ddd(ddd):
    url_ddd = f'https://brasilapi.com.br/api/ddd/v1/{ddd}'
    response = requests.get(url_ddd)
    
    if response.status_code == 200:
        response = response.json()
        df_ddd = pd.DataFrame(response)
        df_ddd['ddd'] = ddd
        return df_ddd
    else:
        alert(1, url_ddd, response.status_code)
        print(f'Falha na obtenção dos dados solicitados. código do erro: {response.status_code}')
```

- Função que cria lista com todos os ddd's existentes no território nacional e posteriomente, utilizando função que cria dataframe, cria uma lista de dataframes a partir de cada ddd nacional e na sequencia os concatena em apenas um dataframe

```python
def dataframe_ddds_nacional():
    ddds = [11, 12, 13, 14, 15, 16, 17, 18, 19, 21, 22, 24, 27, 28, 31, 32, 33, 34, 35, 
            37, 38, 41, 42, 43, 44, 45, 46, 47, 48, 49, 51, 53, 54, 55, 61, 62, 63, 64, 
            65, 66, 67, 68, 69, 71, 73, 74, 75, 77, 79, 81, 82, 83, 84, 85, 86, 87, 88, 
            89, 91, 92, 93, 94, 95, 96, 97, 98, 99]

    dfs = []
    for ddd in ddds:
        df = dataframe_completa_ddd(ddd)
        if df is not None:
            dfs.append(df)
            
    df_ddd_nacional = pd.concat(dfs, ignore_index=True)
    return df_ddd_nacional
```

-Reorganização do dataframe, ordenando as colunas por 'ddd', 'state' e 'cities' e, posteriormente alterando as chaves para estado e cidades. Realizada aplicação de sort_values organizando primariamente por ddd e secundariemente por cidades.

```python
df_nacional_ddd = dataframe_ddds_nacional()

nova_ordem_colunas = ['ddd','state','cities']

df_ddd_nacional = df_nacional_ddd.reindex(columns=nova_ordem_colunas)
df_ddd_nacional.columns = ['ddd', 'estado', 'cidades']
df_ddd_nacional.sort_values(['ddd','cidades'])
```

```python
def df_para_csv(dataframes, nomes_csv):
    for dataframe, nome_csv in zip(dataframes, nomes_csv):
        dataframe.to_csv(f'{nome_csv}.csv', index=False)
        print(f'DataFrame salvo como {nome_csv}.csv com sucesso!')
```

```python
def upload_to_db(df, db_name='../datas_core.db', table_name='nome_tabela'):
    conn = sql.connect(db_name)
    
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    conn.close()
    print(f"Dados carregados para {db_name} na tabela {table_name}")
```

```python
for dataframe, nome_tabela in zip(dataframes, nomes_csv):
    upload_to_db(dataframe, db_name= '../datas_core.db', table_name= nome_tabela)
```

```python
def load_from_db(query, db_name = '../datas_core.db'):
    conn = sql.connect(db_name)
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df
```
